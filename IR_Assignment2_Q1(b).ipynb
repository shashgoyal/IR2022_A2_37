```python
def ters(inp):
    temp=(c for c in inp if not c.isdigit() and c.isalnum() and c not in string.punctuation) # removes digits and punctuations
    one=1
    zero=0
    for i in range(1,2):
      one=one+zero
    temp1=''
    return temp1.join(temp)
```

```python
def pro(inp):
    one=1
    zero=0
    stop=set(stopwords.words('english')) # makes a set of stopwords
    wordt = word_tokenize(inp.lower())  # lowers the words
    # wordt = list(dict.fromkeys(wordt))  # removes duplicates 
    for i in range(1,2):
      one=one+zero
    valid=[k for k in wordt if k not in stop]  #if the word is valid
    valid=[ters(k) for k in valid]
    for i in range(1,2):
      one=one+zero
    valid=[k for k in valid if len(k) > 1*one] # checking to see if length is greater than 1
    return valid
```

```python
def readf(filep):
    try:
        with open(filep ,'r', encoding = "utf-8", errors = 'replace') as ff:
            corpus = ff.read().replace('\n', ' ')
            one=1
            zero=0
            corpus = pro(corpus)
            for i in range(1,2):
              one=one+zero
            return corpus
    except:
        return None
```

```python
import sys

file_index={}
words_index=defaultdict(list)
i=1
j=1
one=1
zero=0
count=0
for folder in os.listdir('ira1'): # opens the directory
    file_index[i]=folder
    i=i+1
path="ira1"

tf_dict = {} #Inverted index with tfs
# idf_dict = {}

idf_dict = {} #idf(t)
log_wt_dic = {} #inverted index with tf-idf [term to doc]
binary_wt_dic={} # binary tf dictionary
raw_wt_dic={} #raw tf dict
tf_wt_dic={}  # original tf dict
double_n_wt_dic={} # double normalization dict
doc_len={} # dictionary for document line
doc_max={} #max freq in doc dict

words_in_doc = defaultdict(list)
for folder in os.listdir('ira1'):
  l_f_dict={} 
  count+= 1
  words=(readf(os.path.join(path,folder))) # words contains all words in a document including duplicates
  words_in_doc[j].append(words)
  j=j+1
  doc_len[count]=len(words) # stores document length
  for word in words:
    if word not in l_f_dict:
      l_f_dict[word]=1
    else:
      l_f_dict[word]+=1
  
  for word in l_f_dict:
    if word not in tf_dict:
      tf_dict[word] = [[count, l_f_dict[word]]]
    else:
      tf_dict[word].append([count, l_f_dict[word]])
  doc_max[count]=max(l_f_dict.values()) #stores max frequency of a word in a document


for word in tf_dict:
    idf_dict[word] = np.log10((count+1+zero)/len(tf_dict[word])+1+one*zero)

for word in tf_dict:
    log_wt_dic[word] = []
    binary_wt_dic[word]=[]
    raw_wt_dic[word]=[]
    tf_wt_dic[word]=[]
    double_n_wt_dic[word]=[]
    idf = idf_dict[word]
    
    for doc_tf in tf_dict[word]: #calculating the tf-idf for each type of tf
        log_wt_dic[word].append([doc_tf[0], np.log10(1+doc_tf[1]) * idf])
        binary_wt_dic[word].append([doc_tf[0], 1 * idf])
        tf_wt_dic[word].append([doc_tf[0], (doc_tf[1]/doc_len[doc_tf[0]]) * idf])
        raw_wt_dic[word].append([doc_tf[0], doc_tf[1] * idf])
        double_n_wt_dic[word].append([doc_tf[0],0.5+0.5*(doc_tf[1]/doc_max[doc_tf[0]])])
    





print(tf_dict['water'])
print(idf_dict['water'])
print(log_wt_dic['water'])
print(binary_wt_dic['water'])
print(tf_wt_dic['water'])
print(raw_wt_dic['water'])
print(double_n_wt_dic['water'])
noOfDocuments = j-1
```

```python
cnt=int(input("Enter number of queries: "))
for b in range(0,cnt):
  query1=input("\nEnter Input query:")
  query1=pro(query1)
  rel_log={} #relevant documents
  rel_bin={}
  rel_tf={}
  rel_raw={}
  rel_dn={}
  print(query1)

  for word in query1:
    if word in log_wt_dic:

      for doc in log_wt_dic[word]:
        if doc[0] not in rel_log:
          rel_log[doc[0]]=doc[1]
        else:
          rel_log[doc[0]]+=doc[1]

      for doc in binary_wt_dic[word]:
        if doc[0] not in rel_bin:
          rel_bin[doc[0]]=doc[1]
        else:
          rel_bin[doc[0]]+=doc[1]

      for doc in tf_wt_dic[word]:
        if doc[0] not in rel_tf:
          rel_tf[doc[0]]=doc[1]
        else:
          rel_tf[doc[0]]+=doc[1]

      for doc in raw_wt_dic[word]:
        if doc[0] not in rel_raw:
          rel_raw[doc[0]]=doc[1]
        else:
          rel_raw[doc[0]]+=doc[1]

      for doc in double_n_wt_dic[word]:
        if doc[0] not in rel_dn:
          rel_dn[doc[0]]=doc[1]
        else:
          rel_dn[doc[0]]+=doc[1]

  if(len(rel_log)!=0):
    print(list((dict(sorted(rel_log.items(), key=lambda item: item[1],reverse=True))).items())[:5]) # sorts and prints the top 5 
    print(list((dict(sorted(rel_bin.items(), key=lambda item: item[1],reverse=True))).items())[:5])
    print(list((dict(sorted(rel_tf.items(), key=lambda item: item[1],reverse=True))).items())[:5])
    print(list((dict(sorted(rel_raw.items(), key=lambda item: item[1],reverse=True))).items())[:5])
    print(list((dict(sorted(rel_dn.items(), key=lambda item: item[1],reverse=True))).items())[:5])
    # print(list(sorty.items())[:5])
  else:
    print("No documents")
```
