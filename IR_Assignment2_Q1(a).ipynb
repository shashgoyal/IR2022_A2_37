```python
import sys

file_index={}
words_index=defaultdict(list)
i=1
j=1
for folder in os.listdir('ira1'): # opens the directory
    file_index[i]=folder
    i=i+1
path="ira1"
words_in_doc = defaultdict(list)
for folder in os.listdir('ira1'):
    words=(readf(os.path.join(path,folder)))
    words_in_doc[j].append(words)
    j=j+1
noOfDocuments = j-1
```
